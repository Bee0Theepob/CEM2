```python
def gcd(a,b):
    if(b==0):
        return int(a)
    return int(gcd(b,a%b))

def pow_mod(a,n,k):
    if(n<=1):
        return int(math.pow(a,n)%k)

    if(n%2==0):
        tmp = pow_mod(a,n//2,k)
        return int((tmp * tmp) % k)
    else:
        tmp = pow_mod(a,n//2,k)
        return int((((tmp * tmp) % k) * a) % k)
    
def mypow(a,n):
    if(n<=1):
        return int(math.pow(a,n))

    if(n%2==0):
        tmp = mypow(a,n//2)
        return int(tmp * tmp)
    else:
        tmp = mypow(a,n//2)
        return int((tmp * tmp) * a)
```

```python
def order_finding1(a,N):
    print("  start order_finding1")
    r=1
    tmp=1
    while(True):
        tmp=(tmp*a)%N
        if(tmp==1):
            print("    found r =",r)
            print("  end order_finding1")
            return r
        r+=1
```

```python
def myshor(N):
    print("N =",N)

    ans0=0
    ans1=0
    iter = 0
    status="failed"
    while(True):
        a = random.randrange(2,N)
        print("attempt",iter,": a =",a)
        k = gcd(a,N)
        if(k != 1):
            ans0=k
            ans1=N//ans0
            status = "founded"
            print(status)
            break
        else:
            r = order_finding1(a,N)
            g = mypow(a,r//2)
            if(r%2==0 and ((g+1)%N!=0) and N%(g+1)):
                ans0=gcd(g+1,N)
                ans1=gcd(g-1,N)
                status = "founded"
                print(status)
                break
        print(status)
        iter+=1
    return ans0,ans1
```

```python
p = 3727
q = 5003
N = p * q
ans0,ans1=myshor(N)
print("--"*10)
print("factor of N =",N)
print("p =",ans0)
print("q =",ans1)
```
